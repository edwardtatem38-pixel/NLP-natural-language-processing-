```python
import pypdf
import io

# The PDF file was uploaded to the Colab environment and its path is known.
# We can access it directly from the /content/ directory.
file_path = '/content/the_nestle_hr_policy_pdf_2012.pdf'

with open(file_path, 'rb') as f:
    pdf_bytes = f.read()

# Open the PDF file from bytes using io.BytesIO
pdf_file = io.BytesIO(pdf_bytes)
pdf_reader = pypdf.PdfReader(pdf_file)

# Extract text from each page
extracted_text = ""
for page_num in range(len(pdf_reader.pages)):
    page = pdf_reader.pages[page_num]
    extracted_text += page.extract_text() + "\n"

# Display the first 1000 characters of the extracted text
print("First 1000 characters of extracted text:")
print(extracted_text[:1000])
print(f"\nTotal characters extracted: {len(extracted_text)}")
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.embeddings.huggingface import HuggingFaceEmbeddings
from langchain.vectorstores import Chroma
from langchain_core.documents import Document # Document is still in langchain_core for this version

# 1. Use your successfully extracted text
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=150)
docs = [Document(page_content=extracted_text)]
chunks = text_splitter.split_documents(docs)

# 2. Re-initialize the embeddings
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
vector_db = Chroma.from_documents(chunks, embeddings)

print("Brain successfully built from extracted text!")
```

```python
# Custom HuggingFaceEmbeddings to bypass problematic langchain.embeddings imports
from langchain_core.embeddings import Embeddings
from sentence_transformers import SentenceTransformer
from typing import List

class CustomHuggingFaceEmbeddings(Embeddings):
    def __init__(self, model_name: str = "all-MiniLM-L6-v2"):
        self.model = SentenceTransformer(model_name)

    def embed_documents(self, texts: List[str]) -> List[List[float]]:
        return self.model.encode(texts).tolist()

    def embed_query(self, text: str) -> List[float]:
        return self.model.encode(text).tolist()

print("CustomHuggingFaceEmbeddings class defined.")
```

```python
import gradio as gr
from langchain.prompts import PromptTemplate
from langchain.schema.runnable import RunnablePassthrough
from langchain.schema.output_parser import StrOutputParser

# 1. Define the RAG Prompt Template
template = """Use the following pieces of context to answer the question at the end.
If you don't know the answer, just say that you don't know, don't try to make up an answer.

{context}

Question: {question}
"""
prompt = PromptTemplate.from_template(template)

# 2. Setup the Custom RAG Chain
# This chain connects the retriever to the LLM via a prompt
qa_chain = (
    {"context": vector_db.as_retriever(search_kwargs={"k": 3}), "question": RunnablePassthrough()}
    | prompt
    | llm
    | StrOutputParser()
)

# 3. Define the Chat Function
def ask_hr(question, history):
    # The custom chain doesn't return source documents directly, only the result.
    # If source documents are desired, the chain structure would need modification.
    response = qa_chain.invoke(question) # Pass question directly to the RunnablePassthrough
    return response

# 4. Launch the Interface
demo = gr.ChatInterface(
    fn=ask_hr,
    title="🏢 Nestlé HR Policy Assistant",
    description="I am an AI assistant trained on the 2012 Nestlé HR Policy. How can I help you today?",
    examples=["What is the policy on maternity leave?", "How do I claim travel expenses?"],
    theme="soft"
)

demo.launch(share=True, debug=True)
```
